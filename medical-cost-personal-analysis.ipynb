# Análise de Dados de Seguros de Saúde

## Introdução

Neste projeto, exploramos um dataset de seguros de saúde que contém informações sobre idade, gênero, índice de massa corporal (BMI), número de filhos, status de fumante, região e custos médicos (charges) para 1338 indivíduos. Nosso objetivo é analisar as correlações entre essas variáveis para obter insights que possam informar a criação de políticas públicas voltadas à saúde.

## Desenvolvimento

### Carregamento e Visualização dos Dados

Primeiramente, carregamos os dados e realizamos uma visualização inicial para entender a estrutura e as características do dataset.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'archive/insurance.csv'
insurance = pd.read_csv(file_path)
insurance.head(1338)
```

Análise Exploratória
Foi realizada uma análise exploratória inicial para entender a distribuição das variáveis principais, como idade, BMI, e custos médicos (charges).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histograma da distribuição de idades
plt.figure(figsize=(10, 6))
sns.histplot(data=insurance, x='age', kde=True, color='skyblue', bins=20)
plt.title('Distribuição de Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

# Gráfico de dispersão com regressão linear para idade e custos de saúde
plt.figure(figsize=(14, 8))
sns.scatterplot(x='age', y='charges', hue='smoker', style='sex', data=insurance, palette='Set1')
sns.regplot(x='age', y='charges', data=insurance, scatter=False, color='grey')
plt.title('Correlação entre Idade, Custos de Saúde, Status de Fumante e Gênero')
plt.xlabel('Idade')
plt.ylabel('Custos de Saúde')
plt.legend(title='Fumante')
plt.show()

# Gráfico de pares (pairplot)
sns.pairplot(insurance, hue='smoker', palette='Set1', markers=['o', 's'])
plt.suptitle('Gráfico de Pares para Idade, Custos de Saúde, Status de Fumante e Gênero', y=1.02)
plt.show()
```

## Distribuição de custos de saúde por faixa etária e status de fumante

```python
plt.figure(figsize=(14, 8))
sns.boxplot(x='age', y='charges', hue='smoker', data=insurance, palette='Set1')
plt.title('Distribuição de Custos de Saúde por Faixa Etária e Status de Fumante')
plt.xlabel('Idade')
plt.ylabel('Custos de Saúde')
plt.legend(title='Fumante')
plt.show()
```

## Correlação entre BMI (Body Mass Index ou Índice de Massa Corporal (IMC) ) e custos de saúde por status de fumante

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=insurance, palette='Set1')
plt.title('Correlação entre BMI e Custos de Saúde por Status de Fumante')
plt.xlabel('BMI')
plt.ylabel('Custos de Saúde')
plt.legend(title='Fumante')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=insurance, x="age", shade=True, color='skyblue')
plt.title('Distribuição de Densidade de Idades')
plt.xlabel('Idade')
plt.ylabel('Densidade')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.scatterplot(x='age', y='charges', hue='smoker', style='sex', data=insurance, palette='Set1')
sns.regplot(x='age', y='charges', data=insurance, scatter=False, color='grey')
plt.title('Correlação entre Idade, Custos de Saúde, Status de Fumante e Gênero')
plt.xlabel('Idade')
plt.ylabel('Custos de Saúde')
plt.legend(title='Fumante')
plt.show()
```

### Principais Observações

Com base no gráfico `pairplot` gerado, aqui estão algumas observações principais que podemos fazer:

#### 1. Idade vs Custos de Saúde
- **Correlação Positiva**: Observa-se uma correlação positiva entre idade e custos de saúde. À medida que a idade aumenta, os custos de saúde tendem a subir.
- **Diferenças entre Fumantes e Não-Fumantes**: Fumantes tendem a ter custos de saúde mais altos em comparação com não-fumantes. Esse padrão é evidente em todas as faixas etárias, mas é particularmente pronunciado em idades mais avançadas.

#### 2. BMI vs Custos de Saúde
- **Custos Aumentam com o BMI**: Há uma tendência clara de aumento nos custos de saúde com o aumento do BMI. Indivíduos com BMI mais alto tendem a incorrer em maiores custos médicos.
- **Impacto do Status de Fumante**: Fumantes com altos valores de BMI apresentam custos de saúde significativamente mais altos do que seus pares não-fumantes. Isso sugere uma interação entre o índice de massa corporal e o status de fumante no aumento dos custos de saúde.

#### 3. Distribuição de Sexo e Custos de Saúde
- **Diferenças de Gênero**: Existe uma diferença nos custos de saúde entre homens e mulheres. Em geral, as mulheres parecem ter custos de saúde mais altos do que os homens.
- **Impacto do Tabagismo nas Mulheres**: As mulheres fumantes tendem a ter custos de saúde mais altos em comparação com as mulheres não-fumantes. Esse padrão é menos evidente nos homens.

### Insights para Políticas Públicas

Com base nas observações acima, algumas sugestões para políticas públicas podem ser feitas:

- **Programas de Prevenção ao Tabagismo**: Dado que os fumantes, especialmente os de BMI mais alto, têm custos de saúde significativamente mais altos, políticas públicas focadas na cessação do tabagismo poderiam ajudar a reduzir esses custos.
- **Intervenções para Idosos**: Como os custos de saúde aumentam com a idade, programas de prevenção e cuidados direcionados à população idosa poderiam ser benéficos.
- **Controle de Peso e Promoção de Estilos de Vida Saudáveis**: A forte correlação entre BMI e custos de saúde sugere que intervenções para controle de peso e promoção de estilos de vida saudáveis poderiam reduzir os custos médicos.
- **Programas Específicos para Mulheres**: Dado que as mulheres, especialmente as fumantes, têm custos de saúde mais altos, políticas direcionadas para a saúde da mulher, incluindo programas de cessação do tabagismo, poderiam ser eficazes.

Essas observações fornecem uma base sólida para a criação de políticas públicas informadas e direcionadas, com o objetivo de reduzir os custos de saúde e melhorar o bem-estar da população.
